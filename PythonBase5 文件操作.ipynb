# 操作流程及模式

## 打开

```python
# 打开 open(file='文件',mode=‘模式’)
# 读写（定位-->读-->写）
# 关闭

'''
r模式： 
    以只读方式打开文件，这是默认模式
    文件指针会放在文件开头
    note:文件不存在会报错
'''
# 1.打开文件
f = open(file = 'FileTest.txt',mode='r') # 相对路径，相对于python源码文件
print(f)
print(type(f))

# 2.读写文件
content = f.read()
print(content)
#f.write('123') # r模式下不能写 UnsupportedOperation: not writable

# 3.关闭文件
f.close()


```

```python

'''
w模式：
    以只写方式打开文件
    文件指针会放在文件开头，所以写入新内容会覆盖
    note:文件不存在会自动创建一个新文件
'''
# 1.打开文件
f = open(file = 'aaw.txt',mode='w') # 没有则创建一个新的


# 2.读写文件
# content = f.read()# r模式下不能读 UnsupportedOperation: not readable
f.write('123')  # 原文件内容被完全覆盖


# 3.关闭文件
f.close()
```

```python
'''
a模式：
    以追加（只写）方式打开文件
    文件指针会放在文件 结尾 ，所以写入新内容会增加大原文件
    note:文件不存在会自动创建一个新文件
'''
# 1.打开文件
f = open(file = 'aaw.txt',mode='a') # 没有则创建一个新的


# 2.读写文件
# content = f.read()# a模式下不能读 UnsupportedOperation: not readable
f.write('abcdef')  # 加到后面


# 3.关闭文件
f.close()
```

```python
'''
增加b：rb,wb,ab
以二进制格式进行操作文件读写(如果是二进制文件则用此项，如音频，图片，视频)
'''
# 1.打开文件，取出内容，获取内容前半部分

# 1.1 打开文件
fromfile = open(file='test.png',mode = 'rb')
#1.2 读取内容
content = fromfile.read()
print(content)
#1.3 关闭文件
fromfile.close()

# 2.打开另一个文件，然后把取出的部分写入到文件2

# 2.1 打开文件
tofile = open(file='test2.jpg',mode = 'wb')
#1.2 读取内容
tofile.write(content[0:len(content)//2])

#1.3 关闭文件
tofile.close()
```

```python
'''
增加+：r+,w+,a+,rb+,wb+,ab+
以读写模式进行打开
'''
# 1.打开文件，取出内容，获取内容前半部分

# 1.1 打开文件 '123456'
file1 = open(file='b.txt',mode = 'r+') # r+不存在则报错

#1.2 读取内容
content = file1.read() # 读写同时进行

print(content)

# 1.3 写入
file1.write('abc') # 'r+'如果没有1.2读取操作，则文件变为abc456,指针在最前面，但不覆盖全部，w覆盖全部
                   #    如果有1.2，先读后写，则写在后面，相当于读完指针在123456后面，再写入


#1.3 关闭文件
file1.close()
```

## 文件操作

### 定位

```python
# f.seek(偏移量,[0,1,2])
# f.tell() 查看指针位置
f = open(file='FileTest.txt',mode = 'r')
print(f.tell())
f.seek(2,0) # 文本文件后面的参数只能为0，1和2只能在带b模式下二进制文件操作
print(f.tell())
print(f.read())
print(f.tell())
help(f.seek)

print('****************')
f = open(file='FileTest.txt',mode = 'rb')
print(f.tell())
f.seek(-2,2) # 在末尾的位置往前偏移2 
print(f.tell())
print(f.read())
print(f.tell())

f.close()
```

### 读

```python
# f.read(字节数) 当前位置开始读取n个
# f.readline([limit])
# f.readlines() 会自动将文件按换行符处理，将处理好的每一行组成一个列表返回
# for in 
# 判断是否可读
# note:
f = open(file='FileTest.txt',mode = 'r')
f.seek(2)
con = f.read(2)
print(con)
print('----',f.tell())
print('***********')

conh = f.readline() # 从当前文件指针位置去读取
print(conh,end='')
print('----',f.tell())

conh = f.readline()
print(conh,end='')
print('----',f.tell())

conh = f.readline(3)
print(conh)
print('----',f.tell())

print('***********')

conhh = f.readlines()
print(conhh)


f.close()
```

```python
# for in 可以直接遍历f本身，也可遍历行列表
import collections
f1 = open(file='FileTest.txt',mode = 'r')

# for i in f1:
#     print(i,end = '')

print(isinstance(f1,collections.Iterator)) # 判断是否是迭代器

cont = f1.readlines()
for l in cont:
    print(l,end='')

f1.close()
```

```python
# 判断当前是否可读功能（增加代码可读性）
f2 = open(file='FileTest.txt',mode = 'w')

if f2.readable(): # 容错处理
    cont = f2.readlines()
    for l in cont:
        print(l,end='')

f2.close()
```

### 写

```python
# f.write('内容') 返回值是写入的字节长度
# 判断是否可写 f.writable()
f3 = open('ww.txt','w')

if f3.writable():
    print(f3.write('abcd'))
    
# f3.write('abcd')仅仅是把‘abc’写入缓冲区，当close时候或其他契机才会写入到文件
f3.flush() # 立即把缓冲区内容刷新写到文件
f3.close()
```
